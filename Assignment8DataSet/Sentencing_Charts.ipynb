```python
import pandas as pd
import numpy as np
import matplotlib
import seaborn as sns
```

```python
sentencing_df = pd.read_csv(r'C:\Users\Administrator\OneDrive\Regis\Ethics\Week 2 - 2\Sentencing.csv')
```

```python
sentencing_df.info()
```

```python
sentencing_df['GENDER'] = sentencing_df['GENDER'].replace('Male name, no gender given', 'Unknown')
sentencing_df['GENDER'] = sentencing_df['GENDER'].replace('Unknown Gender', 'Unknown')
gender_df = sentencing_df.groupby(['GENDER']).mean().reset_index()
gender_df
```

```python
gender_df.plot.bar(x = 'GENDER', y = 'CHARGE_COUNT', title = "Average Charge Count by Gender")
```

```python
gender_df.plot.bar(x = 'GENDER', y = 'LENGTH_OF_CASE_in_Days', title = "Average Charge Count by Gender")
```

```python
gender_df.plot.bar(x = 'GENDER', y = 'AGE_AT_INCIDENT', title = "Average Age by Gender")
```

```python
race_df = sentencing_df.groupby(['RACE']).mean().reset_index()
race_df
```

```python
race_df.plot.bar(x = 'RACE', y = 'CHARGE_COUNT', title = "Average Charge Count by Race")
```

```python
race_df.plot.bar(x = 'RACE', y = 'LENGTH_OF_CASE_in_Days', title = "Average Length of Case by Race")
```

```python
race_df.plot.bar(x = 'RACE', y = 'AGE_AT_INCIDENT', title = "Average Age by Race")
```

```python
gender_crime_df = sentencing_df[['OFFENSE_CATEGORY', 'GENDER', 'AGE_AT_INCIDENT', 'LENGTH_OF_CASE_in_Days']]
```

```python
sns.boxplot(data = gender_crime_df, x = 'GENDER', y = 'AGE_AT_INCIDENT')
```

```python
matplotlib.pyplot.rcParams['figure.figsize'] = [20, 30]
sns.boxplot(data = gender_crime_df, x = 'AGE_AT_INCIDENT', y = 'OFFENSE_CATEGORY', hue = 'GENDER')
```
