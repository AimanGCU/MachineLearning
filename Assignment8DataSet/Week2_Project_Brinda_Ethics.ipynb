```python
# Get only the object columns

obj_cols = list(Diversion_Data.columns)

obj_cols

 

#Import LabelEncoder

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder() 

#Create df to populate with encoded values

encoded_obj_df = pd.DataFrame(columns=obj_cols)

#Apply the transformer

for col in obj_cols:

  encoded_obj_df[col] = le.fit_transform(Diversion_Data[col])


encoded_obj_df.head(10)
```
